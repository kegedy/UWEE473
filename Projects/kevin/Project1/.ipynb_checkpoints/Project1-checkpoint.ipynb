```python
# Imports
import os
import sys
import cmath
import math
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import sympy as sp
from sympy.utilities.lambdify import lambdify
from scipy import signal
%matplotlib inline
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from matplotlib.ticker import LogLocator
import control
```

```python
VOUT = pd.read_csv('/home/kegedy/Documents/EE473/Project1/bandgap_5V_VOUT.csv')
PTAT = pd.read_csv('/home/kegedy/Documents/EE473/Project1/bandgap_5V_PTAT.csv')
PTAT_DERIV = pd.read_csv('/home/kegedy/Documents/EE473/Project1/bandgap_5V_PTAT_deriv.csv')
CTAT = pd.read_csv('/home/kegedy/Documents/EE473/Project1/bandgap_5V_CTAT.csv')
CTAT_DERIV = pd.read_csv('/home/kegedy/Documents/EE473/Project1/bandgap_5V_CTAT_deriv.csv')
```

```python
np.mean(VOUT['/VOUT Y'])
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,12))

ax[0].set_title(f'Voltage Reference')
ax[0].plot(VOUT['/VOUT X'],VOUT['/VOUT Y'],label='Vref')
ax[0].plot(VOUT['/VOUT X'],max(VOUT['/VOUT Y'])*np.ones(len(VOUT['/VOUT Y'])),color='tab:red',linestyle='--',label=f"max={round(max(VOUT['/VOUT Y']),4)} V")
ax[0].plot(VOUT['/VOUT X'],min(VOUT['/VOUT Y'])*np.ones(len(VOUT['/VOUT Y'])),color='tab:red',linestyle='--',label=f"min={round(min(VOUT['/VOUT Y']),4)} V")
ax[0].set_ylabel('Voltage [V]')
ax[0].grid(which='both', axis='both')
ax[0].legend();

ax[1].set_title(r'$\Delta V_{out}/ V_{out}$')
ax[1].plot(VOUT['/VOUT X'],(VOUT['/VOUT Y']-np.mean(VOUT['/VOUT Y']))/np.mean(VOUT['/VOUT Y']),label=r'$\Delta V_{out}/ V_{out}$')
ax[1].set_ylabel('V / V')
ax[1].grid(which='both', axis='both')
ax[1].legend();
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,12))

ax[0].set_title('CTAT Voltage Measurement')
ax[0].plot(CTAT['/net6 X'],CTAT['/net6 Y'],label='CTAT')
ax[0].set_ylabel('Voltage [V]')
ax[0].grid(which='both', axis='both')
ax[0].legend()

ax[1].set_title('CTAT Change with Respect to Temperature')
ax[1].plot(CTAT_DERIV['CTAT_deriv X'],CTAT_DERIV['CTAT_deriv Y'],label='deriv(CTAT)',color='tab:red')
ax[1].set_ylabel('dV / dT')
ax[1].set_xlabel('Time [s]')
ax[1].grid(which='both', axis='both')
ax[1].legend()
plt.show();
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,12))

ax[0].set_title('PTAT Voltage Measurement')
ax[0].plot(PTAT['PTAT X'],PTAT['PTAT Y'],label='PTAT')
ax[0].set_ylabel('Voltage [V]')
ax[0].grid(which='both', axis='both')
ax[0].legend()

ax[1].set_title('PTAT Change with Respect to Temperature')
ax[1].plot(PTAT_DERIV['PTAT_deriv X'],PTAT_DERIV['PTAT_deriv Y'],label='deriv(PTAT)',color='tab:red')
ax[1].set_ylabel('dV / dT')
ax[1].set_xlabel('Time [s]')
ax[1].grid(which='both', axis='both')
ax[1].legend()
plt.show();
```
